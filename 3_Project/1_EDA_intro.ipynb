```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head(3)
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
# df_DA_US
df_plot = df_DA_US['job_location'].value_counts(ascending=False).head(10).reset_index()
df_plot['job_location']  #['count']

sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='dark:b_r',legend=False)
sns.despine()

# plt.barh(df_plot['job_location'],df_plot['count'])
# plt.gca().invert_yaxis()

plt.title('number of jobs per job title')
plt.xlabel('number of jobs')
plt.ylabel('')

plt.show()
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
# df_DA_US
df_plot = df_DA_US['company_name'].value_counts(ascending=False).head(10).to_frame()
df_plot

sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='company_name',hue='count',palette='dark:b_r',legend=False)
sns.despine()

plt.title('number of jobs per job title')
plt.xlabel('number of jobs')
plt.ylabel('')

plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}
# dict_column['job_work_from_home']
# print(len(dict_column))

fig, ax = plt.subplots(1,len(dict_column),figsize=(12,5))
for i,(key,value) in enumerate(dict_column.items()):
    counts = df[key].value_counts()
    din = counts.values
    labels = counts.index

    ax[i].pie(din, labels=labels,startangle=90,autopct='%.1f%%',explode=[0 if label=='False' else 0.1 for label in labels])
    ax[i].set_title(value)

plt.tight_layout()
plt.show()

```

```python
job_filter = ['job_work_from_home','job_no_degree_mention','job_health_insurance']

fig, ax = plt.subplots(1,len(job_filter),figsize=(10,5))

for i,filter in enumerate(job_filter):
    counts = df[filter].value_counts()
    din = counts.values
    labels = counts.index

    ax[i].pie(din,labels=labels,autopct='%.1f%%',startangle=90,explode=[0 if label==False else 0.1 for label in labels])
    ax[i].set_title(filter)


plt.tight_layout()
plt.show()
```
