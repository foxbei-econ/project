```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.head(3)
```

```python
# def time2str(date_in):
#     return date_in.strftime('%Y-%m-%d')

# df['job_posted_date_str'] = df['job_posted_date'].apply(time2str)
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda date_in: date_in.strftime('%Y-%m-%d'))
df[['job_posted_date','job_posted_date_str']].head()
```

```python
from datetime import datetime

# df['days_since_posted'] = df['job_posted_date'] - datetime.now()
current_date = datetime.now()
df['days_since_posted'] = df['job_posted_date'].apply(lambda x: (current_date - x).days)

df[['days_since_posted','job_posted_date']].head()
```

```python
def salary_cat(salary):
    if salary<60_000:
        category = 'Low'
    elif salary>100_000:
        category = 'High'
    else:
        category = 'Medium'
    return category
```

```python
# df_filtered = df[df['salary_year_avg'].notna()].copy()
df_filtered = df.dropna(subset=['salary_year_avg']).copy()
# type(df_filtered)
# df_filtered['salary_category'] = df_filtered['salary_year_avg'].apply(salary_cat)
df_filtered['salary_category'] = df_filtered['salary_year_avg'].apply(lambda salary: 'Low' if salary<60000 else 'High' if salary>100000 else 'Medium')

df_filtered[['salary_year_avg','salary_category']].head()
```
